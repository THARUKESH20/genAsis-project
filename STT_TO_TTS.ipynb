```python
import torch
import librosa
import numpy as np
import IPython.display as ipd
from transformers import AutoProcessor, AutoModelForCTC
from jiwer import wer, cer
from gtts import gTTS  # Google Text-to-Speech
import tempfile  # For storing generated speech

# ---- Load the Tamil STT model ----
model_name = "Amrrs/wav2vec2-large-xlsr-53-tamil"
processor = AutoProcessor.from_pretrained(model_name)
model = AutoModelForCTC.from_pretrained(model_name)

# ---- Load and preprocess the Tamil audio file ----
audio_path = "/content/sample_3.wav"  # Replace with your actual audio file

# Load the audio with Librosa
speech, sample_rate = librosa.load(audio_path, sr=16000)

# Normalize audio (avoid extreme amplitudes)
speech = librosa.util.normalize(speech)

# Remove silent sections (to avoid unwanted noise affecting STT)
speech_trimmed, _ = librosa.effects.trim(speech, top_db=20)

# Process input for model
input_values = processor(speech_trimmed, sampling_rate=16000, return_tensors="pt").input_values

# Get model prediction
with torch.no_grad():
    logits = model(input_values).logits

# Decode output to text
predicted_ids = torch.argmax(logits, dim=-1)
transcription = processor.batch_decode(predicted_ids)[0]

# ---- Post-processing for improved accuracy ----
def clean_tamil_text(text):
    """Removes unwanted spaces, fixes common STT errors, and normalizes Tamil text."""
    text = text.replace("  ", " ")  # Remove extra spaces
    text = text.replace(".", "")  # Remove unnecessary punctuation
    return text.strip()

# Clean transcription output
refined_text = clean_tamil_text(transcription)

# ---- Print Transcriptions ----
print("🔹 Original Tamil Transcription:", transcription)
print("✅ Refined Tamil Text:", refined_text)

# ---- Play the Original Tamil Audio ----
print("🎵 Playing Original Tamil Audio...")
ipd.display(ipd.Audio(audio_path))  # Plays original input audio

# ---- Convert Transcription to Speech (TTS) ----
tts = gTTS(refined_text, lang="ta")  # Tamil TTS
tts_audio_file = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3").name
tts.save(tts_audio_file)

# ---- Play the Generated Tamil Speech ----
print("🔊 Playing Generated Tamil Speech...")
ipd.display(ipd.Audio(tts_audio_file))  # Plays the generated speech
```
